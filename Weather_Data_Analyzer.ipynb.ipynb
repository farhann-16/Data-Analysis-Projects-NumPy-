# üå°Ô∏è Weather Data Analyzer (NumPy Project)
Analyze 7-day temperature data for multiple cities using NumPy features:
- 2D Arrays
- Axis-based operations
- Broadcasting (Heatwave simulation)
- Average, Min, Max, Standard Deviation


```python
import numpy as np
```

```python
# 2D array of temperatures for 4 cities over 7 days
temperatures = np.array([
    [32, 34, 35, 33, 31, 30, 29],   # City A
    [28, 30, 29, 27, 26, 25, 24],   # City B
    [35, 36, 37, 35, 34, 33, 32],   # City C
    [22, 23, 21, 20, 19, 18, 17]    # City D
])
```

```python
city_avg = np.mean(temperatures, axis=1)
day_avg = np.mean(temperatures, axis=0)

print("üìä Average Temperature Per City:", city_avg)
print("üìÜ Average Temperature Per Day:", day_avg)
```

```python
city_min = np.min(temperatures, axis=1)
city_max = np.max(temperatures, axis=1)
city_std = np.std(temperatures, axis=1)

for i in range(len(temperatures)):
    print(f"City {i}: Avg={city_avg[i]:.1f}¬∞C | Min={city_min[i]}¬∞C | Max={city_max[i]}¬∞C | Std Dev={city_std[i]:.2f}")
```

```python
# Add 2¬∞C to simulate a heatwave
heatwave = temperatures + 2
print("üåû Temperatures after Heatwave:\n", heatwave)
```

```python
hottest_city = np.argmax(city_avg)
print(f"üî• City {hottest_city} has the highest average temperature: {city_avg[hottest_city]:.1f}¬∞C")
```

```python
import matplotlib.pyplot as plt

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
cities = ['City A', 'City B', 'City C', 'City D']

plt.figure(figsize=(10, 6))
for i in range(len(temperatures)):
    plt.plot(days, temperatures[i], label=cities[i], marker='o')

plt.title("üìà Temperature Trends for 4 Cities (7 Days)")
plt.xlabel("Days")
plt.ylabel("Temperature (¬∞C)")
plt.legend()
plt.grid(True)
plt.tight_layout()
# plt.savefig("temperature_trends.png")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(cities, city_avg, color=['orange', 'blue', 'red', 'green'])

plt.title("üìä Average Temperature per City")
plt.ylabel("Temperature (¬∞C)")
plt.tight_layout()
plt.savefig("average_temperature_per_city.png")
plt.show()
```
