# ðŸ’¼ E-Commerce Sales Data Simulator & Analyzer (NumPy Project)

This project simulates and analyzes weekly sales data of an online store across multiple product categories using NumPy.

### âœ… Key Features
- 2D arrays for sales data
- Axis-based operations (sum, mean)
- Broadcasting to calculate revenue and profit
- Slicing for weekday/weekend comparison


```python
import numpy as np
```

```python
np.random.seed(42)

categories = ['Electronics', 'Clothing', 'Books', 'Toys', 'Groceries']

# Units sold for 7 days Ã— 5 categories
units_sold = np.random.randint(20, 100, size=(7, 5))

# Unit price per category
unit_prices = np.array([1500, 800, 300, 500, 200])

# Profit margin per category
profit_margins = np.array([0.15, 0.30, 0.25, 0.20, 0.10])
```

```python
# Revenue = Units Ã— Price
revenue = units_sold * unit_prices

# Profit = Revenue Ã— Margin
profit = revenue * profit_margins
```

```python
units_per_day = np.sum(units_sold, axis=1)
units_per_category = np.sum(units_sold, axis=0)

daily_revenue = np.sum(revenue, axis=1)
revenue_per_category = np.sum(revenue, axis=0)

avg_units_category = np.mean(units_sold, axis=0)

highest_selling = categories[np.argmax(units_per_category)]
```

```python
# Assume: 0 = Monday ... 6 = Sunday
weekday_units = np.sum(units_sold[:5], axis=0)
weekend_units = np.sum(units_sold[5:], axis=0)
```

```python
print("ðŸ“Š Total Units Sold Per Category:", dict(zip(categories, units_per_category)))
print("ðŸ’µ Revenue Per Category:", dict(zip(categories, revenue_per_category)))
print("ðŸ“ˆ Average Units Per Category:", dict(zip(categories, avg_units_category.round(2))))
print("ðŸ”¥ Highest Selling Category:", highest_selling)

print("\nðŸ“… Units Sold (Weekdays):", dict(zip(categories, weekday_units)))
print("ðŸ“† Units Sold (Weekend):", dict(zip(categories, weekend_units)))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(categories, revenue_per_category, color='skyblue')
plt.title("Revenue by Product Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.savefig("revenue_by_category.png")
plt.grid(True)
plt.show()
```
